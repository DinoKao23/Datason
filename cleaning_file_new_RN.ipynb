```python
# import packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import merged data csv
df = pd.read_csv("C:\\Users\\riley\\OneDrive\\LMU Datathon Spring 2024\\New folder\\append\\All Years\\merged.csv")
```

```python
# Convert 'date_column' to datetime if it's not already in datetime format
df['period_of_performance_start_date'] = pd.to_datetime(df['period_of_performance_start_date'],errors='coerce')

# Create a mask to filter rows where the date is past October 2018
mask = df['period_of_performance_start_date'] >= '2018-10-1'

# Apply the mask to the DataFrame to keep only the rows after October 1, 2018
filtered_df = df[mask]
```

```python
# Convert 'action_date' to datetime if it's not already in datetime format
filtered_df['action_date'] = pd.to_datetime(filtered_df['action_date'])

# Sort the DataFrame by 'contract_id' and 'action_date'
sorted_df = filtered_df.sort_values(by=['contract_award_unique_key', 'action_date'], ascending = [True,False])

# Keep oldest action date for each contract id, removing duplicates of contracts
sorted_df.drop_duplicates(subset='contract_award_unique_key', inplace=True)
```

```python
# Create indicator for small business that is minority or women owned
sorted_df['new_column'] = (sorted_df['contracting_officers_determination_of_business_size'] == 'SMALL BUSINESS') & ((sorted_df['woman_owned_business'] == 't') | (sorted_df['minority_owned_business'] == 't'))
```

```python
# Create a mask to filter rows where the business is women or minority owned and small
mask = sorted_df['new_column'] == 1

# Apply the mask to the DataFrame to keep only the rows after October 1, 2018
new_filtered_df = sorted_df[mask]
```

```python
# Convert 'start_date' and 'end_date' to datetime if they are not already in datetime format
new_filtered_df['period_of_performance_start_date'] = pd.to_datetime(new_filtered_df['period_of_performance_start_date'])
new_filtered_df['period_of_performance_current_end_date'] = pd.to_datetime(new_filtered_df['period_of_performance_current_end_date'])

# Calculate the number of months between 'start_date' and 'end_date'
new_filtered_df['contract_length'] = ((new_filtered_df['period_of_performance_current_end_date'] - new_filtered_df['period_of_performance_start_date']).dt.days / 30)
```

```python
# Convert 'start_date' and 'end_date' to datetime if they are not already in datetime format
new_filtered_df['period_of_performance_start_date'] = pd.to_datetime(new_filtered_df['period_of_performance_start_date'])
new_filtered_df['ordering_period_end_date'] = pd.to_datetime(new_filtered_df['ordering_period_end_date'])

# Calculate the number of months between 'start_date' and 'end_date'
new_filtered_df['contract_length_other'] = ((new_filtered_df['ordering_period_end_date'] - new_filtered_df['period_of_performance_start_date']).dt.days / 30)
```

```python
# Using numpy.where to replace values
new_filtered_df['contract_length'] = np.where((new_filtered_df['contract_length'] <= 0 | new_filtered_df['contract_length'].isnull()), new_filtered_df['contract_length_other'], new_filtered_df['contract_length'])
#new_filtered_df['contract_length'] = np.where(new_filtered_df['contract_length'].isnull(), new_filtered_df['contract_length_other'], new_filtered_df['contract_length'])
```

```python
# Count null values in 'your_column'
null_count = new_filtered_df['contract_length'].isnull().sum()

# Count values below zero in 'your_column'
below_zero_count = (new_filtered_df['contract_length'] < 0).sum()

# Display the results
print(f"Null count: {null_count}")
print(f"Below zero count: {below_zero_count}")
```
